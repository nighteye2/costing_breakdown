```python
import matplotlib.pyplot as plt
import seaborn as sns
import math


PO_cost = int(input('Enter PO cost:'))
material_cost = int(input('Enter material cost:'))
labor_cost = int(input('Enter labor cost:'))
Travelling_cost = int(input('Enter Travelling cost'))
additional_cost = int(input('Enter additional cost like food etc'))
material_transport_cost = int(input('Enter material transport cost'))
vendor_po = int(input('Enter vendor po'))
fittings_cost = int(input('Enter fittings cost'))


total_cost = material_cost + labor_cost + Travelling_cost + additional_cost + material_transport_cost + vendor_po
profit = PO_cost - total_cost
material_cost_percent = (material_cost/PO_cost)*100
labor_cost_percent = (labor_cost/PO_cost)*100
Travelling_cost_percent = (Travelling_cost/PO_cost)*100
additional_cost_percent = (additional_cost/PO_cost)*100
material_transport_cost_percent = (material_transport_cost/PO_cost)*100
vendor_po_percent = (vendor_po/PO_cost)*100
fittings_cost_percent = (fittings_cost/PO_cost)*100
profit_percent = (profit/PO_cost)*100
total_cost_percent = (total_cost/PO_cost)*100


plt.figure(figsize=(9,5))
data = [material_cost_percent,labor_cost_percent,Travelling_cost_percent,additional_cost_percent,profit_percent,material_transport_cost_percent,vendor_po_percent,fittings_cost_percent]
labels = ['material_cost','labor_cost','Travelling_cost','additional_cost','profit','material_transport_cost','vendor_cost','fittings_cost']
plt.pie(data,labels=labels,autopct='%1.1f%%')
plt.title('costing breakdown')
print('\n')
print(f'PO amount sent on this project is: {PO_cost}')
print('\n')
print(f'Total cost on this project is: {total_cost}')
print(f'Total percent of cost on this project is: {total_cost_percent}')
print('\n')
print(f'Total profit on this project is: {profit}')
print(f'Total percent of profit on this project is: {profit_percent}')
```
